```python
# 행렬곱 
# 행렬 A, B, C 가 주어졌을 때 이 규칙이 올바르게 적용되는지 확인해보자

import numpy as np

A = np.arange(0, 31, 2).reshape(4, 4)
B = np.arange(1, 32, 2).reshape(4, 4)
C = np.arange(2, 33, 2).reshape(4, 4)

print(A)
print()
print(B)
print()
print(C)
```

```python
# AB != BA
print(A @ B)
print()
print(B @ A)
```

```python
# A(B + C) == AB + AC
print(A @ (B + C))
print()
print((A @ B) + (A @ C))
```

```python
# (A + B)C = AC + BC
print((A + B) @ C)
print()
print((A @ C) + (B @ C))
```

```python
# (AB)T = BT AT
print((A @ B).T)
print()
print(B.T @ A.T)
```

```python
# 실습 1
# 1 - 100의 정수로 (10, 10) 행렬을 생성한 후,
# np.random.shuffle() 을 이용하여 행렬을 뒤섞은 다음, 행렬을 정렬하시오

a = np.arange(1, 101).reshape(10, 10)
print(a)
print()
np.random.shuffle(a)
print('shuffle')
print(a)
print()

print('sort')
print(np.sort(a))
```

```python
# 실습 2
# 1 - 30 의 정수로 (3, 10) 행렬을 생성한 후,
# np.random.shuffle() 을 이용하여 행렬을 섞은 다음 axis 값을 바꾸어 정렬하시오

b = np.arange(1, 31).reshape(3, 10)
print(b)
print()

np.random.shuffle(b)
print('shuffle')
print(b)
print()

print('axis = 1')
print(np.sort(b, axis = 1))
print()
print('axis = 0')
print(np.sort(b, axis = 0))
```
