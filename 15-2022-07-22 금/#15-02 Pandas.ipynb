```python
import numpy as np
import pandas as pd

# 데이터 
d1 = pd.date_range('20220701', periods = 100)

df3 = pd.DataFrame(np.random.randn(100, 4), index = d1, columns = list('ABCD'))

df3.head()
```

```python
# Viewing Data 

# 축(column)별로 정렬
df3.sort_index(axis = 1, ascending = False)

df3.sort_index(axis = 0, ascending = False)
```

```python
# Selection - Getting 1

df3.A     # A index 출력

df3['A']  # A index 출력

df3[['A', 'B']] # A, B index 출력
```

```python
# Selection - Getting 2

df3[0:3]  # 0 - 3 index 

df3['20220701': '20220710'] # 2022-07-01 ~ 2022-07-10 까지

df3[['A', 'B']][0:3] # A, B Column 과 0 - 3 index 부분 출력
```

```python
# Selection by Label 1

# 라벨을 사용하여 횡단면을 획득

d1[0] # Timestamp('2022-07-01 00:00:00', freq='D')

df3.loc[d1[0]]
```

```python
# Selection by Label 2

# 라벨을 사용하여 여러 축(의 데이터) 을 얻는 방법

df3.loc[:, ['A', 'B']] # 전체 index 와 A, B Column

df3.loc['20220701': '20220710', ['A', 'B']] # 2022-07-01 ~ 2022-07-10 index 와 A, B Column

df3.loc['20220701', ['A', 'B']] # 2022-07-01 index 와 A, B Column
```

```python
# Selection by Position 1

# 넘겨 받은 정수의 위치로 선택

df3.iloc[3] # 3번째 index

df3.iloc[3:5, 0:2] # 3 - 5 index, 0 - 2 Column

df3.iloc[[1, 2, 4], [0, 2]]  # ---- S.18 Weiter
```

```python
# Selection by Position 2

df3.iloc[1:3, :]

df3.iloc[:, 1:3]
```

```python
# Selection by Position 3

df3.iloc[1, 1]

df3.iat[1, 1]
```

```python
# Boolean Indexing 1

df3.A > 0

df3[df3.A > 0]
```

```python
# Boolean Indexing 2
df3 > 0

df3[df3 > 0]
```

```python
# Boolean Indexing 3

df4 = df3.copy()

from numpy import random
r = random.randint(1000, size = 100)

df4['E'] = r
df4

r1 = list(range(300, 401))
df4['E'].isin(r1)

df4[df4['E'].isin(r1)]
```

```python
# Setting 1

# 새로운 열을 날짜와 인덱스에 따라 자동적으로 줄이 맞춰지도록 설정
l = list(range(1001, 1101))

s1 = pd.Series(l, index = pd.date_range('20220701', periods = 100))

s1

df3['F'] = s1

df3
```

```python
# Setting 2

# 라벨에 따라 값을 설정

df3.at[d1[0], 'A'] = 0

df3
```

```python
# Setting 3

# 위치에 따라 값을 설정

df3.iat[0, 1] = 0

df3
```

```python
# Setting 4

# numpy 배열로 할당

df3.loc[:, 'D'] = np.array([5] * len(df3))

df3
```

```python
# Setting 5

df4 = df3.copy()
df4 > 0

df3

df4[df4 > 0] = -df4

df4
```

```python
# Missing Data 1

# reindexing 으로 지정된 축을 변경 / 추가 / 삭제 할 수 있음
# reindexing 은 데이터 복사본을 반환

df5 = df3.reindex(index = d1[0:4], columns = list(df3.columns) + ['E'])
df5

df5.loc[d1[0]: d1[1], 'E'] = 1
df5
```

```python
# Missing Data 2

# Missing Data를 갖고 있는 행들을 삭제

df5.dropna(how = 'any')
```

```python
# Missing Data 3

# Missing Data를 채움
df5.fillna(value = 5)
```

```python
# Missing Data 4

# 값이 nan인 boolean mask를 얻음

pd.isna(df5)
```

```python
# Operation - Stats 1

# 일반적으로 연산은 결측치를 제외
#서술적인 통계를 수행

df3.mean()

df3.mean(1)
```
