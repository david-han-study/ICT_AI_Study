```python
# 지도 학습
# 데이터 준비
from sklearn.datasets import load_digits
digits = load_digits() # digit.data, digit.target 사용
```

```python
# 데이터 분할
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix

X_train, X_test, y_train, y_test = train_test_split(digits.data,
                                                    digits.target,
                                                    test_size = 0.33)
```

```python
# 모델 준비
from sklearn.neighbors import KNeighborsClassifier
m = KNeighborsClassifier(n_neighbors = 10, weights = 'distance')

# 학습
m_train = m.fit(X_train, y_train)
```

```python
# 예측
# result = neigh_10_train.predict(X_test)
y_est = m.predict(X_test)

print(classification_report(y_test, y_est))
print(confusion_matrix(y_test, y_est))

# print(result)
```
