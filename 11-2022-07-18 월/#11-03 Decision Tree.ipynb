```python
# Decision Tree 실습 - Weather 데이터에 적용하기

# 1. 데이터 준비
import pandas as pd
data = pd.read_csv('weather.csv', sep=',')

data
```

```python
# 2. 카테고리 문자열을 수치로 변환

# mapping dictionary 만들기
outlook_dic = {
               'overcast' : 0,
               'rainy' : 1,
               'sunny' : 2
               }
temperature_dic = {
                  'cool' : 0,
                  'hot' : 1,
                  'mild' : 2
                  }
humidity_dic = {
                'high' : 0,
                'normal' : 1
                }
windy_dic = { False : 0,
              True : 1 }

# dictionary 를 활용해 데이터 매핑
data['outlook'] = data['outlook'].map(outlook_dic)
data['temperature'] = data['temperature'].map(temperature_dic)
data['humidity'] = data['humidity'].map(humidity_dic)
data['windy'] = data['windy'].map(windy_dic)

data
```

```python
# 3. 결정 트리 분류기 인스턴스 생성

# Decision Tree classifier instance 생성
from sklearn import tree
clf = tree.DecisionTreeClassifier()
clf
```

```python
# 4. 결정 트리 분류기 학습
a = clf.fit(data.iloc[:, :4], data['play'])
```

```python
# 5. 오늘같은 날씨에는 필드에 나갈까? 예측하기

# 맑고, 온도가 적당하고, 습도가 높고, 바람이 분다
a = clf.predict([[2,2,0,1]])
print(a)

# 비가 오고, 온도가 적당하고, 습도가 높고, 바람이 분다
b = clf.predict([[1,2,0,1]])
print(b)
```
