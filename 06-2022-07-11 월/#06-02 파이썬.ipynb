```python
# sys 모듈
# os 모듈: 운영체제와 관련된 기능 가진 모듈

import os

print(os.name)            # 현재 운영체제
print(os.getcwd())        # 현재 폴더
print(os.listdir())       # 현재 폴더 내부 요소 리스트
print(os.system('ls'))    # 시스템 명령어 실행

with open('a.txt', 'w') as f:
  f.write('hallo')
os.rename('a.txt', 'A.txt') # 파일 이름 변경

os.remove('A.txt')          # 파일 삭제
```
