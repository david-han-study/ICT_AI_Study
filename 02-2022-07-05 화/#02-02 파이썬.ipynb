```python
# 조건문
# 짝수 홀수 구분
num = input('숫자 입력하세여')
last_num = int(num[-1])

if last_num == 0 or last_num == 2 or last_num == 4 or last_num == 6 or last_num == 8:
  print('짝수')

num = input('숫자 입력하세여')
last_num = num[-1]

if last_num in "24680":
  print('짝수')

num = int(input('숫자 입력하세여'))

if num % 2 == 0:
  print('짝수')
```

```python
# 조건문이 오로지 두 가지로만 구분될 때 if else 사용
num = int(input('숫자 입력하세요'))

if num % 2 == 0 :
  print('짝수')
else : 
  print('홀수')
```

```python
# 리스트
# 다른 언어와 차이점: 자료형에 상관없이 들어갈 수 있다.

list_a = [234, 12, 25, '문자', True, False]

# 인덱스: 대괄호 안에 들어간 숫자
print(list_a[2])

# 리스트 슬라이스
print(list_a[1:4]) 

# 새 값으로 변경할 수 있다
list_a[0] = '변경된 문자'
print(list_a[0])

# 인덱스에 음수 넣어 뒤에서부터 요소 선택하기
print(list_a[-1])

# 리스트 접근 연산자를 이중으로 사용할 수 있다.
print(list_a[3][1]) # list_a[3] : '문자' 의 [1] 번째 글자인 자 를 출력

# 리스트안에 리스트
list_b = [[1,2,3], [4,5,6]]
print(list_b[1])
print(list_b[1][2])
```

```python
# 리스트 연산자: +, *
list_c = [1, 2, 3]
list_d = [4, 5, 6]
print(list_c + list_d)
print(list_c * 3)
```
