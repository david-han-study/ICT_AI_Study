```python
# 리스트 함수 

# len() : 리스트의 개수 
list_a = [1, 2, 3, 4]
print(len(list_a))
```

```python
# append() : 리스트에 요소 추가
list_a.append(5)
print(list_a)
```

```python
# insert() : 리스트 중간에 추가
list_b = [1, 3, 4]
list_b.insert(1, 2) # insert(위치, 요소)
print(list_b)
```

```python
# extend() : 원래 리스트 뒤에 새로운 리스트의 요소 모두 추가 / 매개변수로 리스트
list_c = [1, 2, 3]
list_c.extend([4, 5, 6])
print(list_c)

# [1, 2, 3] + [4, 5, 6] 과 결과는 같다
# extend 는 원본이 바뀜: list_c => [1, 2, 3, 4, 5, 6]
# [1, 2, 3] + [4, 5, 6]은 원본이 바뀌지 않음 => list_c 는 [1, 2, 3]
```

```python
# 리스트 요소 제거하기

# del 리스트명[인덱스]
# 리스트명.pop(인덱스)

list_d = [1, 2, 3, 4, 5, 6]

del list_d[0]
print(list_d)

list_d = [1, 2, 3, 4, 5, 6]
list_d.pop(2)
print(list_d)
```

```python
# pop은 한번에 하나만 삭제 가능
# del은 한번에 여러개 삭제 가능

list_e = [1, 2, 3, 4, 5, 6]
del list_e[2:4]
print(list_e)

list_e = [1, 2, 3, 4, 5, 6]
del list_e[:3]
print(list_e)
```

```python
# remove() : 값으로 제거하기 => 특정 값을 지정
list_f = [1, 2, 3, 4, 5, 6]
list_f.remove(2)

print(list_f)

list_g = [1, 1, 1, 2, 3, 4]
list_g.remove(1) # 같은 값이 있을 땐 제일 먼저 값을 삭제

print(list_g)
```

```python
# clear() : 모두 제거하기
list_h = [1, 2, 3, 4, 5, 6]
list_h.clear()

print(list_h)
```

```python
# in 연산자 : 특정 값이 리스트 내부에 있는지 확인
list_i = [1, 2, 3, 4, 5, 6]

1 in list_i # True return

0 not in list_i # True
```
