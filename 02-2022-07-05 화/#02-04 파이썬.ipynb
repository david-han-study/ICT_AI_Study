```python
# 딕셔너리: 키를 기반으로 값을 저장하는 것

a = {
    "key1": 10,   # 문자열을 키로 사용하기
     1: "일",     # 숫자를 키로 사용하기
     False: 30    # boolean을 키로 사용하기
}
print(a[False])
```

```python
# 딕셔너리에 값 추가 하기
# 딕셔너리[키] = 새로운 값

b = {
    "name": "abcd",
     "age": 10
}
b["name"] = 'asdf'

# 없는 키를 사용하면 새로운 키가 추가
b['ort'] = 'seoul'

print(b)
```

```python
# 딕셔너리 요소 제거
# del

c = {
    "name": "abcd",
    "age": 10
}

del c['name']
print(c)
```

```python
# 딕셔너리 예제

d = {}

d['name'] = '이름'
d['age'] = 20

print(d)
```

```python
# 딕셔너리 내부에 키가 있는지 확인하기

# get()
e = {
    "name": "abcd",
    "age": 10
}

v = e.get('name')
print(v)

v2 = e.get('asdf') # 없는 키를 get 하면 None 리턴
print(v2)

# in 연산자
if 'nameㅁ' in e :
  print('name 있음')
```

```python
# for와 딕셔너리의 조합

# for 키 변수 in 딕셔너리:
#   코드

f = {
    "name": "abcd",
    "age": 10
}

for i in f:         # f의 키들만 i 에 들어감
  print(i)          # 키만 출력됨
  print(f[i])       # 값이 출력됨
```
