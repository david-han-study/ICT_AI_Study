```python
# Boolean
# True | False 값만 가질 수 있다.
# 비교 연산자를 통해 만들수 있음
print(10 == 100)
print("가방" == "가방")

# 문자열의 비교에서는 '사전'을 생각하면 된다
print("가방" < "하마")
```

```python
# Boolean 끼리 논리 연산자 사용 가능
# not 
# and
# or
print(not True)
```

```python
# 조건문
# if 조건 :
#   들여쓰기 해야함
a = 4
if a == 4:
  print('ok')
```

```python
# 날짜/시간 출력하기
# datetime
import datetime

now = datetime.datetime.now()

print(now)
print(now.year)
print(now.month)
print(now.day)
print(now.hour)
print(now.minute)
print(now.second)
```

```python
# 계절 출력
import datetime
# now = datetime.datetime.now()
month = int(input('계절을 입력하세요 '))

if 3 <= month <= 5:
  print('{}월은 봄'.format(month))
if 6 <= month <= 8:
  print('{}월은 여름'.format(month))
if 9 <= month <= 11:
  print('{}월은 가을'.format(month))
if month == 12 or 1 <= month <= 2:
  print('{}월은 겨울'.format(month))
```
