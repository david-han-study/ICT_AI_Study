```python
# 객체
# 객체를 만드는 함수

def create_student(name, korean, math, english, science):
    return {
        'name': name,
        'korean': korean,
        'math': math,
        'english': english,
        'science': science
    }

students = [
    create_student('윤인성', 80, 60, 70, 30),
    create_student('김철수', 80, 60, 70, 30),
    create_student('박영수', 80, 60, 70, 30),
    create_student('윤아린', 80, 60, 70, 30),
    create_student('김영희', 80, 60, 70, 30),
    create_student('이연희', 80, 60, 70, 30)
]

print('이름\t총점\t평균')
for s in students: 
    sum = s['korean'] + s['math'] + s['english'] + s['science']
    average = sum / 4
    print(s['name'], sum, average, sep='\t')

def get_sum(student):
    return student['korean'] + student['math'] + student['english'] + student['science']

def get_average(student):
    return get_sum(student) / 4

def get_string(student):
    return f'{get_sum(student)} {get_average(student)}'
```

```python
# __str__() 
# 클래스에서 __str__() 함수 정의 하면 str() 함수 호출할 때 __str__() 함수가 자동으로 호출
class Student:
  def __init__(self, name, korean, math, english, science):
      self.name = name
      self.korean = korean
      self.math = math
      self.english = english
      self.science = science
  
  def get_sum(self):
      return self.korean + self.math + self.english + self.science

  def get_average(self):
      return self.get_sum() / 4

  def __str__(self):
    return '{}\t{}\t{}'.format(
      self.name,
      self.get_sum(),
      self.get_average()
    )
  
  def __eq__(self, value):
    return self.get_sum() == value.get_sum()

  def __ne__(self, value):
    return self.get_sum() != value.get_sum()
        
  def __gt__(self, value):
    return self.get_sum() > value.get_sum()
      
  def __ge__(self, value):
    return self.get_sum() >= value.get_sum()

  def __lt__(self, value):
    return self.get_sum() < value.get_sum()    

  def __le__(self, value):
    return self.get_sum() <= value.get_sum()    
  
students = [
  Student('윤인성', 80, 60, 70, 30),
  Student('김철수', 80, 60, 70, 30),
  Student('박영수', 80, 60, 70, 30),
  Student('윤아린', 80, 60, 70, 30),
  Student('김영희', 80, 60, 70, 30),
  Student('이연희', 80, 60, 70, 30)
]

student_a = Student('윤인성', 80, 60, 70, 30)
student_b = Student('김철수', 80, 60, 70, 30)
print(str(student_a))
print(student_a == student_b)
print(student_a != student_b)
print(student_a > student_b)
print(student_a >= student_b)
print(student_a < student_b)
print(student_a <= student_b)
  
```
