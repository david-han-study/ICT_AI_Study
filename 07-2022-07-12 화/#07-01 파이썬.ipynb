```python
# __init__.py
# - 패키지 읽을 때 어떤 처리를 수행해야 하거나 
# 패키지 내부의 모듈들을 한꺼번에 가져오고 싶을 때 사용
# 패키지 읽어들일 때 __init__.py 를 가장 먼저 실행한다
```

```python
!mkdir test_package
```

```python
%%writefile test_package/module_a.py
a = 'a모듈 변수'
```

```python
%%writefile test_package/module_b.py
b = 'b모듈 변수'
```

```python
%%writefile test_package/__init__.py
__all__ = ['module_a', 'module_b']

print('test_package를 읽었습니다')
```

```python
from test_package import *

print(module_a.a)
print(module_b.b)
```
