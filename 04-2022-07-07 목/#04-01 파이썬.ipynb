```python
# 함수

# def 함수 이름(매개변수) :
#   문장
def print_n(n, value): 
  for i in range(n):
    print(value)

# 가변 매개 변수
# 매개변수를 원하는 만큼 받을수 있는 함수
# - 가변 매개변수 뒤에는 일반 매개변수 올 수 없음
# - 가변 매개변수는 하나만 사용할 수 있음

# def 함수 이름(매개변수, 매개변수, *가변 매개변수):
#   문장
def print_values(n, *values):
  for i in range(n):
    for j in values: 
      print(j) 

print_n(2, '안녕')
print_values(3, '안녕', '메롱')
```

```python
# 기본 매개변수
# - 매개변수 값 입력하지 않았을 경우 매개변수에 들어가는 기본값

def print_default(value, n=2):
  for i in range(n):
    print(value)

print_default('안녕')
print_default('안녕', 4)

# 키워드 매개변수
# 매개변수 이름을 지정해서 입력하는 매개변수
print_default('안녕', n=4)
```

```python
# return

def sum_all(start, end): 
  result = 0 

  for i in range(start, end + 1):
    result += i
  
  return result

sum_all(1, 100)
```
