```python
# 기본 매개변수와 키워드 매개변수를 활용
def sum_all(start=0, end=100, step=1):
  result = 0

  for i in range(start, end + 1, step):
    result += i
  
  return result

sum_all(1, 100, 5)
```

```python
# 문제
# 매개변수로 전달된 값들을 모두 곱해서 리턴하는 가변 매개변수 함수를 만들어보세요.
def mul(*values): 
  result = 1
  
  for value in values:
    result *= value
  return result

print(mul(5, 7, 9, 10))
```

```python
# 피보나치 수열

def fibonacci(n): 
  if n == 1:
    return 1
  if n == 2:
    return 1

  else :
    return fibonacci(n - 1) + fibonacci(n - 2)

print(fibonacci(10))
```

```python
# 메모화
# - 재귀 함수를 사용하면서 코드가 빠르게 실행되게
# 같은 값을 한 번만 계산하도록 코드를 수정
# 딕셔너리를 사용해서 한 번 계산한 값을 저장
# 처리 수행하지 않고 메모된 값 돌려주면서 코드 속도 향상

d = {
    1: 1,
    2: 2
}

def fibonacci2(n):
  if n in d:
    return d[n]
  else: 
    result = fibonacci2(n - 1) + fibonacci2(n - 2)
    d[n] = result
    return result

print(fibonacci2(20))
```
