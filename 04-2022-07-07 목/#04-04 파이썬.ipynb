```python
# 파일처리 실습

# 랜덤하게 1000명의 키와 몸무게 만들기

import random 

h = list('가나다라마사아자차카타파하') # 문자열을 리스트로

with open('random_info.txt', 'w') as file:
  for i in range(1000):
    name = random.choice(h) + random.choice(h)
    weight = random.randrange(40,100)
    height = random.randrange(150, 200)

    file.write(f"{name}, {weight}, {height}\n")
```
