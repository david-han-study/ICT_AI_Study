```python
# try 구문 내부에서 return 키워드를 사용해야 하는 경우
# return 후에도 finally는 실행한다

def test():
  print('일단 1번 시작')
  try:
    print('try 1번')
    return
    print('이건 실행 안될듯')
  except:
    pass
  else:
    print('else')
  finally:
    print('finally')

test()
```

```python
# 반복문 중에서도 break 키워드 후에 finally 구문 실행

print('시작')
while True:
  try:
    print('try')
    break
  except:
    pass
  finally:
    print('finally')
print('종료')
```

```python
# 예외 객체 (exception object)
# - 예외 발생 시 예외 정보가 저장되는 곳
# try:
#   예외가 발생할 가능성이 있는 구문
# except 예외의 종류 as 예외 객체를 활용할 변수 이름:
#   예외가 발생했을 떄 실행할 구문

list_num = [12, 34, 5, 61, 66]

try:
  number_input = int(input('정수 입력: '))
  print(f'{number_input}번째 요소: {list_num[number_input]}')
except Exception as e:
  print('e: ', e)
```

```python
# 예외 구분하기
# except를 여러개 사용할 수 있음
list_num = [12, 34, 5, 61, 66]

try:
  number_input_b = int(input('정수 입력: '))
  print(f'{number_input_b}번째 요소: {list_num[number_input_b]}')
except ValueError:
  print('정수를 입력하세여')
except IndexError:
  print('해당 인덱스가 없어여')
```
