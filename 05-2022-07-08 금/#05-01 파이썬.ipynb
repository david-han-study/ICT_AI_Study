```python
# 오류와 예외

# 구문 오류 (Syntax error): 프로그램 실행 전에 발생하는 오류
# 런타임 오류 (Runtime error) / 예외 (exception) : 프로그램 실행 중에 발생하는 오류

# 예외 처리
# - 조건문을 사용하는 방법
# - try 구문을 사용하는 방법

# 조건문으로 예외 처리하기

# isdigit() : 숫자로만 구성된 글자인지 확인
input_a = input('정수 입력')

# 사용자 입력이 숫자로만 되어 있을 때
if input_a.isdigit():
  input_a = int(input_a)
  print(input_a)
else: 
  print('정수를 입력하지 않았습니다')
```

```python
# try except 구문
# - 예외 처리할 수 있는 구문
# try:
#   에러가 발생할 가능성이 있는 코드
# except:
#   예외가 발생했을 때 실행할 코드

try:
  input_b = int(input('정수 입력'))

  print(input_b)
except:
  print('에러 입니다')
```

```python
# except 에 pass 조합하면 에러가 발생해도 넘어간다

list_a = ['12', '1235', '3456', '바보', '멍청이', '456']
list_number_1 = []

for item in list_a:
  try:
    float(item) # float으로 변환이 안되면 에러 -> pass
    list_number_1.append(item)
  except:
    pass

print(list_number_1)
```

```python
# try except else 구문
# try: 
#   예외가 발생할 가능성이 있는 코드
# except:
#   예외가 발생했을 때 실행할 코드
# else:
#   예외가 발생하지 않았을 때 실행할 코드

try:
  number_input_b = int(input('정수 입력'))
except:
  print('정수가 아니에요')
else:
  print('입력한 숫자: ', number_input_b)
```

```python
# finally 구문

# try: 
#   예외가 발생할 가능성이 있는 코드
# except:
#   예외가 발생했을 때 실행할 코드
# else:
#   예외가 발생하지 않았을 때 실행할 코드
# finally:
#   무조건 실행할 코드

try:
  number_input_b = int(input('정수 입력'))
except:
  print('정수가 아니에요')
else:
  print('입력한 숫자: ', number_input_b)
finally:
  print('어쨋든 끝났어요')
```
