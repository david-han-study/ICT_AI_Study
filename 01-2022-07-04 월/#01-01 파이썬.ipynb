```python
# 이스케이프 문자: \n, \t
print('"\n안녕"')
```

```python
print('\thello')
```

```python
# 여러 줄 문자열
print("""동해물과 백두산이 마르고 닳도록
하느님이 보우하사 우리 나라 만세
무궁화 삼천리 화려강산
대한사람 대한으로 길이 보전하세
""")
```

```python
# 문자열 연산자 -> 문자 + 숫자는 안됨
print('안녕' + '하세유')
```

```python
# 문자열 반복 연산자: *
print('안녕' * 10)
print(10 * '안녕') # 이것도 됨
```

```python
# 문자 선택 연산자(인덱싱): []
# 문자열 내부의 문자 하나를 선택
# 대괄호 안에 선택할 문자의 위치를 지정
print('abcd'[0])

# 대괄호 안 숫자를 음수로 입력도 가능
#  a  b  c  d  e
# -5 -4 -3 -2 -1
# => -1은 마지막 글자
```

```python
# 문자열 범위 선택 연산자(슬라이싱): [ : ]
# 문자열의 특정 범위를 선택
# 마지막 숫자를 포함하지 않음
# 슬라이싱을 해도 원본이 바뀌진 않는다

print('abcdef'[2:5])
```

```python
# 콜론의 앞을 생략하면 문자열의 맨 앞부터
# 콜론의 뒤를 생략하면 n부터 맨 뒤까지
print('abcdef'[:4])
print('abcdef'[4:])
```

```python
# 문자열의 길이
# len()
print(len('abcd'))
```
