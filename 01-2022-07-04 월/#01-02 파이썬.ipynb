정수형
*   소수점이 없는 숫자
*   0, 1, 243, -51
*   정수(integer)

실수형
* 소수점이 있는 숫자
* 0.0, 23.13
* 실수(floating)



```python
# 숫자

# 정수형
# 소수점이 없는 숫자
# 0, 1, 243, -51
# 정수(integer)


# 실수형
# 소수점이 있는 숫자
# 0.0, 23.13
# 실수(floating)

print(1234)
```

```python
# 정수 나누기 연산자: //
# -> 몫만 구한다
print(5//2) # 2.5 의 2만 나옴
```

```python
# 정수 나누기 연산자: %
# 나머지만 구한다
print(5%2)
```

```python
# 제곱 연산자: **
print(2**3)
```

```python
# 연산자 우선순위
# 수학과 같지만 프로그래밍에서는 정확하게 괄호로 묶어주자
```

```python
# TypeError 예외
# => 서로 다른 자료를 연산할 경우
```

```python
# 복합 대입 연산자
# += : 덧셈 후 대입
a = 10
a += 10
print(a)
```

```python
# 문자열 복합 대입 연산자
# +=, *= 만 사용 가능
string = '안녕하세여'
string += '!'
print(string)
string *= 2 # string = string * 2
print(string)
```

```python
# 사용자 입력: input()
# input 은 문자열로 입력 받는다
a = input('입력하세여') # '입력하세요' => prompt / 입력받고 a 변수에 할당
```

```python
# 캐스트(cast)
# int() : 정수형(int) 로 변환
# float() : 실수(float)으로 변환
a = int('52')
print(type(a))
```

```python
# str() : 숫자를 문자열로 변환
a = str(123)
print(type(a))
```
