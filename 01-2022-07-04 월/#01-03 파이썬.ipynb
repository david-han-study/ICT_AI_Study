```python
# format() 
# 중괄호 포함한 문자열 뒤에 마침표 찍고 format() 함수 사용하되, 
# 중괄호 개수와 format 함수 안 매개변수의 개수는 반드시 같아야함
# 문자열의 중괄호 기호가 format() 함수 괄호 안의 매개변수로 차례로 대치되면서 숫자가 문자열이 됨
format_a = "{}만원".format(10)
print(format_a)
```

```python
# 정수
ouput_a = "{:d}".format(52)
print(ouput_a)

# 특정 칸에 출력하기
ouput_b = "{:5d}".format(52)
ouput_c = "{:10d}".format(52)
print(ouput_b)
print(ouput_c)

# 빈칸을 0으로 채우기
ouput_d = "{:05d}".format(52)
ouput_e = "{:05d}".format(-52)
print(ouput_d)
print(ouput_e)
```

```python
# 기호와 함께 출력하기
ouput_f = "{:+d}".format(52) # 양수
ouput_g = "{:+d}".format(-52) # 음수
print(ouput_f)
print(ouput_g)
print()
ouput_h = "{: d}".format(52) # 양수: 기호 부분 공백
ouput_i = "{: d}".format(-52) # 음수: 기호 부분 공백
print(ouput_h)
print(ouput_i)
print()
# 조합하기
ouput_j = "{:+5d}".format(52) # 기호를 뒤로 밀기: 양수
ouput_k = "{:+5d}".format(-52) # 기호를 뒤로 밀기: 음수
ouput_l = "{:=+5d}".format(52) # 기호를 앞으로 밀기: 양수
ouput_m = "{:=+5d}".format(-52) # 기호를 앞으로 밀기: 음수
ouput_n = "{:+05d}".format(52) # 0으로 채우기: 양수
ouput_o = "{:+05d}".format(-52) # 0으로 채우기: 음수
print(ouput_j)
print(ouput_k)
print(ouput_l)
print(ouput_m)
print(ouput_n)
print(ouput_o)
```

```python
# 부동소수점 출력
output_a = "{:f}".format(52.273)
output_b = "{:15f}".format(52.273) # 15칸 만들기
output_c = "{:+15f}".format(52.273) # 15칸에 부호 추가하기
output_d = "{:+015f}".format(52.273) # 15칸에 부호 추가하고 0으로 채우기

print(output_a)
print(output_b)
print(output_c)
print(output_d)
```

```python
# 의미없는 소수점 제거하기
ouput_a = 52.0
ouput_b = "{:g}".format(ouput_a)
print(ouput_a)
print(ouput_b)
```

```python
# upper() : 문자의 알파벳을 대문자로 바꿈
# lower() : 문자의 알파벳을 소문자로
# strip() : 문자열 양옆의 공백을 제거
# lstrip() : 왼쪽의 공백을 제거
# rstrip() : 오른쪽의 공백을 제거

input_a = """
  안녕하세요
문자열의 함수를 알아봅시다
"""
print(input_a)
print()
print(input_a.strip())
```

```python
# 문자열 구성 파악하기 => true, false 반환
# isalnum() : 알파벳 or 숫자 
# isalpha() : 알파벳
# isidentifier() : 식별자로 사용가능?
# isdecimal() : 정수?
# isdigit() : 숫자로 인식 가능?
```

```python
# find() : 왼쪽부터 찾아서 처음 등장하는 위치 찾음
# rfind() : 오른쪽부터 찾아서 처음 등장하는 위치 찾음
output_a = "안녕안녕하세요".find("안녕")
print(output_a)
output_b = "안녕안녕하세요".rfind("안녕")
print(output_b)
```

```python
# in 연산자 : 문자열이 있는지 확인할 때 사용 => true, false 반환
print('안녕' in ' 안녕하세요')
```

```python
# split() 함수 : 문자열을 특정한 문자로 자름 => 리스트로 반환
a = "10 20 30 40 50".split(" ")
print(a)
```
